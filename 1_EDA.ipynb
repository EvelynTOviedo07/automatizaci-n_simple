### 4. ¿Cuál es la correlación entre el ingreso anual y la puntuación de gasto?
- **Objetivo**: Identificar si existe una relación lineal entre el ingreso y el gasto.
- **Respuesta esperada**: Una gráfica de dispersión y el cálculo del coeficiente de correlación entre Annual Income (k$) y Spending Score (1-100).

```python
# 1. Calcular el coeficiente de correlación de Pearson
correlacion = df['Annual Income (k$)'].corr(df['Spending Score (1-100)'])
print(f"El coeficiente de correlación de Pearson es: {correlacion:.4f}")

# 2. Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# 3. Crear el gráfico de dispersión
sns.scatterplot(
    data=df, 
    x='Annual Income (k$)', 
    y='Spending Score (1-100)', 
    s=80, 
    color='darkviolet', 
    alpha=0.7,
    edgecolor='black'
)

# 4. Personalizar el gráfico y añadir el valor de correlación al título
plt.title(f'Ingreso Anual vs. Puntuación de Gasto\n(Correlación de Pearson: {correlacion:.2f})', 
          fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Ingreso Anual (k$)', fontsize=12)
plt.ylabel('Puntuación de Gasto (1-100)', fontsize=12)

# Mostrar la gráfica
plt.show()
```

### 5. ¿Cómo varía la puntuación de gasto en diferentes grupos de ingresos anuales?
- **Objetivo**: Examinar cómo los clientes en diferentes rangos de ingresos se comportan en términos de gasto.
- **Respuesta esperada**: Una gráfica de cajas o de violín que muestre la Spending Score (1-100) para diferentes rangos de Annual Income (k$).

```python
# 1. Crear los rangos de ingresos anuales (Bajo, Medio, Alto)
# Usamos los percentiles o cortes lógicos del dataset habitual (ej. <40k, 40k-70k, >70k)
bins_income = [0, 40, 70, 140]
labels_income = ['Ingresos Bajos (<40k)', 'Ingresos Medios (40k-70k)', 'Ingresos Altos (>70k)']
df['Income_Group'] = pd.cut(df['Annual Income (k$)'], bins=bins_income, labels=labels_income, include_lowest=True)

# 2. Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# 3. Crear el gráfico de violín
sns.violinplot(
    data=df, 
    x='Income_Group', 
    y='Spending Score (1-100)', 
    palette='Set3', 
    hue='Income_Group', 
    legend=False,
    inner='quartile' # Muestra las líneas de los cuartiles y la mediana por dentro
)

# 4. Personalizar el gráfico
plt.title('Distribución de Puntuación de Gasto por Rango de Ingresos', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Rango de Ingreso Anual (k$)', fontsize=12)
plt.ylabel('Puntuación de Gasto (1-100)', fontsize=12)

# Mostrar la gráfica
plt.show()
```

### 7. ¿Qué grupos de edad gastan más en promedio?
- **Objetivo**: Identificar los grupos de edad que tienen una mayor puntuación de gasto en promedio.
- **Respuesta esperada**: Una gráfica de barras que compare la puntuación de gasto promedio entre diferentes grupos de edad.

```python
# 1. Crear los rangos de edad (si no los has creado antes)
bins = [18, 30, 45, 60, 70]
labels = ['18-30', '31-45', '46-60', '61-70']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)

# 2. Calcular el promedio de Spending Score por cada grupo de edad
# Agrupamos por 'Age_Group' y extraemos la media de la puntuación de gasto
df_grouped = df.groupby('Age_Group', observed=False)['Spending Score (1-100)'].mean().reset_index()

# 3. Configurar el tamaño de la figura
plt.figure(figsize=(9, 5))

# 4. Crear el gráfico de barras
ax = sns.barplot(
    data=df_grouped, 
    x='Age_Group', 
    y='Spending Score (1-100)', 
    palette='Blues_r', # Paleta de colores degradados (del más oscuro al más claro)
    hue='Age_Group',
    legend=False
)

# 5. Añadir las etiquetas de los valores encima de cada barra para mayor claridad
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=3, fontsize=10, fontweight='bold')

# 6. Personalizar títulos y etiquetas
plt.title('Puntuación de Gasto Promedio por Rango de Edad', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Rango de Edad', fontsize=12)
plt.ylabel('Puntaje de Gasto Promedio (1-100)', fontsize=12)
plt.ylim(0, 100) # Mantener la escala completa del puntaje para no distorsionar la percepción

# Mostrar la gráfica
plt.show()
```

### 9. ¿Cuál es la distribución conjunta de la edad y el ingreso anual?
- **Objetivo**: Entender cómo se distribuyen estas dos variables en conjunto.
- **Respuesta esperada**: Una gráfica de dispersión con una densidad de puntos o un gráfico de hexágonos que muestre la distribución conjunta de Age y Annual Income (k$).

```python
# Crear el gráfico de distribución conjunta usando hexágonos
grid = sns.jointplot(
    data=df, 
    x='Age', 
    y='Annual Income (k$)', 
    kind='hex', 
    cmap='Blues',        # Paleta de colores (más oscuro = más clientes)
    height=7,            # Tamaño del gráfico (cuadrado)
    gridsize=15,         # Tamaño de los hexágonos (menos = más grandes)
    edgecolor='gray',    # Línea sutil para separar los hexágonos
    linewidth=0.5
)

# Ajustar los títulos (ya que jointplot maneja su propia estructura de figuras)
grid.fig.suptitle('Distribución Conjunta: Edad vs. Ingreso Anual', fontsize=14, fontweight='bold', y=1.02)
grid.set_axis_labels('Edad (Años)', 'Ingreso Anual (k$)', fontsize=12)

# Mostrar la gráfica
plt.show()# tu código
```

### 10. ¿Cómo se distribuyen los clientes en función de la puntuación de gasto y el género?
- **Objetivo**: Analizar la relación entre la puntuación de gasto y el género.
- **Respuesta esperada**: Una gráfica de dispersión o un gráfico de violín que muestre la Spending Score (1-100) separada por género.

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# 1. Crear el gráfico de violín base
sns.violinplot(
    data=df, 
    x='Gender', 
    y='Spending Score (1-100)', 
    palette='Pastel2', 
    hue='Gender',
    legend=False,
    inner='quartile', # Muestra la mediana y los cuartiles
    alpha=0.7
)

# 2. Agregar los puntos individuales encima para ver la distribución real
sns.stripplot(
    data=df, 
    x='Gender', 
    y='Spending Score (1-100)', 
    color='black', 
    alpha=0.3,       # Puntos semi-transparentes para no saturar
    jitter=0.15      # Dispersión horizontal sutil para que no se encimen
)

# 3. Personalizar el gráfico
plt.title('Distribución de Puntuación de Gasto por Género', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Género', fontsize=12)
plt.ylabel('Puntuación de Gasto (1-100)', fontsize=12)

# Mostrar la gráfica
plt.show()
```
